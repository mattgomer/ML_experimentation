```python
#add description. I made this to subtract noise from my phone's astronomy picture


__author__="Matt Gomer"
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
mpl.rc('image', cmap='gray')
import scipy
from scipy import signal
from scipy import fftpack
from PIL import Image
import seaborn as sns
import time
import copy
import sys
```

```python
source_image=mpimg.imread('../images/phone_astrophoto.jpg')
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
fig, ax = plt.subplots(figsize=(15, 10))

bw_image=rgb2gray(source_image)
ax.imshow(bw_image)
plt.show()
sns.kdeplot(bw_image[0:100,0:100].flatten())
```
