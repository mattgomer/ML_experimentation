```python
#I made this notebook to subtract noise from my phone's astronomy picture. 
#Right now it is just a simple technique which sets everything below a threshold to zero
#I should experiment with more advanced techniques in the future

#need to try again with raw file instead of jpg

__author__="Matt Gomer"
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
mpl.rc('image', cmap='gray')
import scipy
from scipy import signal
from scipy import fftpack
from PIL import Image
import seaborn as sns
import time
import copy
import sys
```

```python
source_image=mpimg.imread('../images/phone_astrophoto.jpg')
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
fig, ax = plt.subplots(figsize=(15, 10))

bw_image=rgb2gray(source_image)
ax.imshow(bw_image)
plt.show()
sns.kdeplot(bw_image[0:100,0:100].flatten())
```

```python
noise_subtract_image=copy.copy(bw_image)
threshold=30
noise_subtract_image=bw_image-threshold

noise_subtract_image[noise_subtract_image<0]=0.1
fig, ax = plt.subplots(2,1,figsize=(15, 15))
# print(np.min(bw_image),np.max(bw_image))
ax[0].matshow(bw_image,vmin=0,vmax=250)
ax[1].matshow(noise_subtract_image,vmin=0,vmax=250)
plt.show()
print('subtracted from red pixels')
fig, ax = plt.subplots(1,1,figsize=(15, 15))
mask=np.zeros(np.shape(noise_subtract_image))
mask[bw_image<threshold]=1
ax.matshow(mask,cmap='bwr')
plt.show()
# plt.hist(mask[0:100,0:100].flatten(),bins=100)
# plt.show()
```
